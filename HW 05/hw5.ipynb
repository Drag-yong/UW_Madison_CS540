```python
    print('Q3b:')
    Y = []
    for i in days:
        Y.append(i)
    Y = np.array(Y)
    print(Y)
```

```python

    # Z = X^T * X
    print('Q3c:')
    Z = np.dot(np.transpose(X), X)
    print(Z)
```

```python
    # I = inverse of Z = (X^T * X)^-1
    print('Q3d:')
    I = np.linalg.inv(Z)
    print(I)
```

```python

    # PI = I * X^T = (X^T * X)^-1 * X^T
    print('Q3e:')
    PI = np.dot(I, np.transpose(X))
    print(PI)
```

```python
    ## hat_beta = PI*Y = ((X^T)*X)^(-1)*(X^T)*Y
    print('Q3f:')
    hat_beta = np.dot(PI, Y)
    print(hat_beta)
```
